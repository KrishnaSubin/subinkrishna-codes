```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(labels="duration",axis=1,inplace=True)
df.drop(labels="default",axis=1,inplace=True)
df.drop(labels="day",axis=1,inplace=True)
df.drop(labels="contact",axis=1,inplace=True)
df.drop(labels="housing",axis=1,inplace=True)
```

```python
df.hist(figsize=(12,8),bins=20)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['balance','campaign','pdays','previous']]=scaler.fit_transform(df[['balance','campaign','pdays','previous']])
```

```python
df
```

```python
from sklearn import preprocessing   
```

```python
label = preprocessing.LabelEncoder()
df[ 'job' ]= label.fit_transform(df[ 'job' ])   
df[ 'marital' ]= label.fit_transform(df[ 'marital' ])
df[ 'education' ]= label.fit_transform(df[ 'education' ])
df[ 'loan' ]= label.fit_transform(df[ 'loan' ])
df[ 'month' ]=label.fit_transform(df[ 'month' ])
df[ 'poutcome' ]= label.fit_transform(df[ 'poutcome' ])
df[ 'deposit' ]= label.fit_transform(df[ 'deposit' ])
```

```python
df
```

```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True,cmap="YlGnBu")
plt.plot()
```

```python
X=df.drop(labels="deposit",axis=1)
y=df['deposit']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.30,random_state = 101)
```

# Decision tree classifer

```python
from sklearn.tree import DecisionTreeClassifier  
classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)  
classifier.fit(X_train, y_train)
pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
recall_bc=recall_score(y_test, pred, average='macro')
print("recall= ",recall_bc)

classi_bc=f1_score(y_test, pred)
print("f1 score= ",classi_bc)

precision_bc=precision_score(y_test, pred, average='macro')
print("precision= ",precision_bc)

```

# Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None)
classifier.fit(X_train, y_train)
pred=classifier.predict(X_test)
print(accuracy_score(y_test,pred))

recall_bc=recall_score(y_test, pred, average='macro')
print("recall= ",recall_bc)

classi_bc=f1_score(y_test, pred)
print("f1 score= ",classi_bc)

precision_bc=precision_score(y_test, pred, average='macro')
print("precision= ",precision_bc)

```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=0)
lr.fit(X_train, y_train)
pred=lr.predict(X_test)
print(accuracy_score(y_test,pred))

recall_bc=recall_score(y_test, pred, average='macro')
print("recall= ",recall_bc)

classi_bc=f1_score(y_test, pred)
print("f1 score= ",classi_bc)

precision_bc=precision_score(y_test, pred, average='macro')
print("precision= ",precision_bc)


```

# Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
eclf2= BaggingClassifier(base_estimator=DecisionTreeClassifier()).fit(X_train, y_train)
var1=eclf2.predict(X_test)

recall_bc=recall_score(y_test, var1, average='macro')
print("recall= ",recall_bc)

classi_bc=f1_score(y_test, var1)
print("f1 score= ",classi_bc)

precision_bc=precision_score(y_test, var1, average='macro')
print("precision= ",precision_bc)
```

```python
# The precision, recall and f1 score ogf bagging classifier is higher than other machine learning models.
```
